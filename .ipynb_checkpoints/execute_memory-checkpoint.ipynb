```python
from autogen_vertex_mcp_system_claude import *
```

```python
"""Example usage"""

# Initialize system
config = SystemConfig(
    autogen_config={
        "config_list": [{
            "model": "claude-opus-4-1-20250805",
            "api_type":"anthropic",
            "api_key":"Claude key"
        }],
        "temperature":0.1,
        "cache_seed": 42
    }
)

executor = MissionExecutor(config)

# Example missions
missions = [
    "Zbuduj pipeline do analizy danych CSV z obsługą błędów",
    "Stwórz system ML z continuous learning i rollback",
    "Zaprojektuj adaptacyjny system przetwarzania z retry logic"
]


result = await executor.execute_mission(missions[0])

print(f"\n{'='*60}")
print(f"Mission: {missions[0]}")
print(f"{'='*60}")
print(f"Plan nodes: {len(result['final_plan'].get('nodes', []))}")
print(f"Plan edges: {len(result['final_plan'].get('edges', []))}")
print(f"Learned from: {len(result['learning_context'].get('analyzed_missions', []))} missions")
print(f"\nFinal Plan:")
print(json.dumps(result['final_plan'], indent=2, ensure_ascii=False))
```
